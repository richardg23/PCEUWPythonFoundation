import pandas as pd
import matplotlib.pyplot as plt

# run plots in the notebook
%matplotlib inline

url = "http://pbpython.com/extras/sample-salesv2.csv"

sales = pd.read_csv(url)

```python
sales.head()
```

#### Rename columns with spaces 

```python
sales.columns = ['acct_num', 'name', 'sku', 'category', 'quantity', 'unit_price', 'ext_price', 'date']
```

```python
sales.head(10)
```

#### Subset the dataframe to contain only the name, category, quantity and unit price columns

```python
shirt_df = sales[['name', 'category', 'quantity', 'unit_price']]
shirt_df.head(10)
```

#### Subset the dataframe to contain only shirt sales

```python
shirt_df = shirt_df[shirt_df['category']== 'Shirt']
shirt_df.head(10)
```

```python
sales.describe()
```

```python
shirt_df.sort_values(by='quantity', ascending=False).head(20)
```

#### Calculate the total cost per shirt sale

```python
shirt_df['shirt_sales'] = shirt_df.quantity * shirt_df.unit_price
```

```python
shirt_df.head()
```

#### Group the shirt sales by company name

```python
shirt_by_company = shirt_df.groupby('name').sum()
```

```python
shirt_by_company.head(8)
```

```python
shirt_by_company = shirt_df.groupby('name', as_index=False).sum()
```

```python
shirt_by_company.head(10)
```

#### Group the top 10 shirt sales

```python
top_shirt_sellers = shirt_by_company.sort_values(by='shirt_sales', ascending=False).head(10)
```

```python
top_shirt_sellers
```

```python
shirt_plot = top_shirt_sellers.plot(kind="bar", 
                             title="Total Shirt Sales by Company",
                             x="name",
                             y="shirt_sales")
shirt_plot.set_xlabel("Company Name")
shirt_plot.set_ylabel("Shirts Sold ($)")
```
